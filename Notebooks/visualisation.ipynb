```python
import matplotlib.pyplot as plt
import folium as fm
import pandas as pd
import table_evaluator as te
```

```python
def plot_map(data):
    map_center = [data['latitude'].mean(), data['longitude'].mean()]
    map = fm.Map(location=map_center, zoom_start=12)

    for _, row in data.iterrows():
        fm.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=1,
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(map)
    return map
```

```python
def plot_evaluation(real, fake, cat=None):
    table = te.TableEvaluator(real, fake, cat)
    table.visual_evaluation()
    return
```

```python
data = pd.read_csv('/workspaces/Synthetic-Data-Generation/Data/Real/dataset_gps.csv')
map = plot_map(data)

plot_evaluation(data, data, cat=None)
```
